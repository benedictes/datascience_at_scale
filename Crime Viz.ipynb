#MAP

```python
SF_COORDINATES = (37.76, -122.45)  
map = fm.Map(location=SF_COORDINATES, zoom_start=12)
```

```python
MAX_RECORDS = 1000
for i, row in sf[0:MAX_RECORDS].iterrows():  
    map.simple_marker(
        location = [row['Y'],row['X']], 
        clustered_marker = True) 
```

```python
map.create_map(path='map.html')
```
